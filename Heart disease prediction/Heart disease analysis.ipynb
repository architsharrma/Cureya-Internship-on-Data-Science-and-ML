# Model Building:

### Using Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
```

```python
x=df.drop(['target','age_group'],axis=1)
y=df['target']
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.score(x,y)
```

Without train test split method the model is giving us accuracy of 85%

Let's check with train test split method:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

Our model is giving us accuracy of 90% which is far better than without implementing it with train test split method.

#### Let's predict manually by entering the data.

```python
model.predict([[63,1,3,145,233,1,0,150,0,2.3,0,0,1]])
```

We are getting output as 1 it means the person is having a heart disease.

```python
pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```

#### With the help of confusion matrix we can see that 6 times our actual value is 1 but model is predicting 0.
